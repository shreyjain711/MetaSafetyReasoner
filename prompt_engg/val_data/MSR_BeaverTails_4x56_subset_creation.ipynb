```python
import pandas as pd
from datasets import load_dataset

# Load the dataset from Hugging Face
dataset = load_dataset("PKU-Alignment/BeaverTails", 'default')

# Convert the dataset to a pandas DataFrame
df = dataset['330k_train'].to_pandas()

# Display the first few rows of the DataFrame
display(df.head())
```

```python
category_df = df['category'].apply(lambda x: pd.Series(x))
df = pd.concat([df, category_df], axis=1)
display(df.head())
```

```python
df.drop(columns=['category'], inplace=True)
display(df.head())
```

```python
# Get unique combinations of the one-hot encoded columns
unique_combinations = df.drop(columns=['prompt', 'response', 'is_safe']).drop_duplicates()

display(unique_combinations)
```

```python
df['true_count'] = df.loc[:, categories].sum(axis=1)
df['true_count'].value_counts()
```

```python
# Create sampling pools based on conditions

# Pool for is_safe==True
safe_pool = df[df['is_safe'] == True].copy()

# Pools for is_safe==False and true_count==1, one for each category
unsafe_count_1_pools = {}
categories = df.columns[3:-1].tolist() # Get the list of category columns
for category in categories:
    unsafe_count_1_pools[category] = df[(df['is_safe'] == False) & (df['true_count'] == 1) & (df[category] == True)].copy()

# Pools for true_count==2, 3, 4, 5
count_2_pool = df[df['true_count'] == 2].copy()
count_3_pool = df[df['true_count'] == 3].copy()
count_4_pool = df[df['true_count'] == 4].copy()
count_5_pool = df[df['true_count'] == 5].copy()

# Pool for true_count>=6
count_ge_6_pool = df[df['true_count'] >= 6].copy()

print("Sampling pools created.")
```

```python
assert len(safe_pool['is_safe'].unique()) == 1 and safe_pool['is_safe'].unique()[0] == True
assert len(count_ge_6_pool['is_safe'].unique()) == 1 and count_ge_6_pool['is_safe'].unique()[0] == False and any(count_ge_6_pool['true_count'].unique()>=6) and count_ge_6_pool['true_count'].unique()[0] >= 6
assert len(count_2_pool['is_safe'].unique()) == 1 and count_2_pool['is_safe'].unique()[0] == False and any(count_2_pool['true_count'].unique()==2) and count_2_pool['true_count'].unique()[0] == 2
assert len(count_3_pool['is_safe'].unique()) == 1 and count_3_pool['is_safe'].unique()[0] == False and any(count_3_pool['true_count'].unique()==3) and count_3_pool['true_count'].unique()[0] == 3
assert len(count_4_pool['is_safe'].unique()) == 1 and count_4_pool['is_safe'].unique()[0] == False and any(count_4_pool['true_count'].unique()==4) and count_4_pool['true_count'].unique()[0] == 4
assert len(count_5_pool['is_safe'].unique()) == 1 and count_5_pool['is_safe'].unique()[0] == False and any(count_5_pool['true_count'].unique()==5) and count_5_pool['true_count'].unique()[0] == 5
```

```python
import random

def sample_and_remove(pool, n_samples):
    """Samples n_samples from the pool and removes them from the pool."""
    if len(pool) < n_samples:
        print(f"Warning: Not enough samples in pool. Requested: {n_samples}, Available: {len(pool)}")
        n_samples = len(pool)
    samples = pool.sample(n_samples)
    pool.drop(samples.index, inplace=True)
    return samples

unique_sets = []
num_sets = 4
samples_per_set = 56

for i in range(num_sets):
    current_set_samples = []

    # 13 is_safe==True samples
    current_set_samples.append(sample_and_remove(safe_pool, 13))

    # 14 is_safe==False, true_count==1, one for each of the 14 categories
    for category in categories:
        current_set_samples.append(sample_and_remove(unsafe_count_1_pools[category], 1))

    # 7 from each of true_count==2,3,4,5
    current_set_samples.append(sample_and_remove(count_2_pool, 7))
    current_set_samples.append(sample_and_remove(count_3_pool, 7))
    current_set_samples.append(sample_and_remove(count_4_pool, 7))
    current_set_samples.append(sample_and_remove(count_5_pool, 7))


    # 1 case from true_count>=6
    current_set_samples.append(sample_and_remove(count_ge_6_pool, 1))

    # Concatenate samples for the current set
    unique_sets.append(pd.concat(current_set_samples))

# Display the first few rows of each generated set and their size
for j, dataset in enumerate(unique_sets):
    print(f"\nSet {j+1} ({len(dataset)} rows):")
    display(dataset.head())
```
